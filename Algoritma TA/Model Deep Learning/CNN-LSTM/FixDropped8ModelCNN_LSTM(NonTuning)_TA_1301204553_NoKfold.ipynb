```python
import pandas as pd
from google.colab import files
```

```python
# Upload file CSV Myocardial dari lokal
csvMI = files.upload()
```

```python
# Upload file CSV Normal dari lokal
csvNormal = files.upload()
```

```python
extracted_feature_MI1 = list(csvMI.keys())[0]

extracted_feature_normal1 = list(csvNormal.keys())[0]
```

```python
df_extracted_feature_MI = pd.read_csv(extracted_feature_MI1)

df_extracted_feature_normal = pd.read_csv(extracted_feature_normal1)
```

**||| Prediction Process |||**

```python
# Gabungkan DataFrame MI dan Normal
df_combined = pd.concat([df_extracted_feature_MI, df_extracted_feature_normal], ignore_index=True)
```

```python
df_combined
```

```python
df_combined.shape
```

```python
# Tambahkan kolom label
df_combined['Label'] = ['MI'] * len(df_extracted_feature_MI) + ['Normal'] * len(df_extracted_feature_normal)
```

```python
df_combined
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_combined['Encoded_Label'] = label_encoder.fit_transform(df_combined['Label'])
```

```python
df_combined
```

**|| Dropped Table |||**

```python
columns_to_drop = [

    'IQR_wavelet',
    'Q3_wavelet',
    'Q1_wavelet',
    'ZCR Mean',

    'Wavelet min',
    'ZCR Min',
    'RMS Max',
    'RMS_Range',
    'RMS Min',

    'RMS Q1',
    'Wavelet Means',
    'Med_wavelet',
    'RMS_var'
]
```

```python
# Dropping columns that exist in the dataframe to avoid KeyErrors
columns_to_drop = [col for col in columns_to_drop if col in df_combined.columns]
```

```python
# Dropping the specified columns
df_combined = df_combined.drop(columns=columns_to_drop)
```

```python
df_combined
```

**|| Split The Table ||**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_combined.drop(['Label', 'Encoded_Label'], axis=1),
    df_combined['Encoded_Label'],
    test_size=0.3,
    random_state=42
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
X_test_scaled.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout, BatchNormalization, Flatten
from tensorflow.keras.regularizers import l1_l2

mymodel = Sequential()
# First Conv1D layer
mymodel.add(Conv1D(filters=16, kernel_size=3, input_shape=(33, 1), activation='relu'))
mymodel.add(BatchNormalization())
mymodel.add(MaxPooling1D(pool_size=2))

# Conv1D layer kedua
mymodel.add(Conv1D(filters=16, kernel_size=3, activation='relu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)))
mymodel.add(BatchNormalization())
mymodel.add(MaxPooling1D(pool_size=2))
mymodel.add(Dropout(0.2))

# LSTM layer pertama
mymodel.add(LSTM(32, activation='relu', return_sequences=True))
mymodel.add(Dropout(0.2))

# LSTM layer kedua
mymodel.add(LSTM(64, activation='relu'))
mymodel.add(Dropout(0.2))

# Dense layer
mymodel.add(Dense(32, activation='relu'))
mymodel.add(Dropout(0.5))

# Output layer
mymodel.add(Dense(1, activation='sigmoid'))

# Compile the model
mymodel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
mymodel.summary()
```

```python
X_train_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))
```

```python
X_train_reshaped.shape
```

```python
X_test_reshaped.shape
```

```python
X_train_reshaped
```

```python
X_test_reshaped
```

```python
epochs = 100
batch_size = 32
```

```python
history = mymodel.fit(X_train_reshaped, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.3)
```

```python
loss, accuracy = mymodel.evaluate(X_test_reshaped, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

```python
lossTrain, accuracyTrain = mymodel.evaluate(X_train_reshaped, y_train)
print(f"Loss: {lossTrain}, Accuracy: {accuracyTrain}")
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions on the test set
y_pred_proba = mymodel.predict(X_test_reshaped)

# Apply threshold to get binary predictions
threshold = 0.5
y_pred = (y_pred_proba > threshold).astype(int)

# Convert predictions to 1D array
y_pred = y_pred.flatten()

# Decode labels
y_test_decoded = label_encoder.inverse_transform(y_test)
y_pred_decoded = label_encoder.inverse_transform(y_pred)

# Compute confusion matrix
cm = confusion_matrix(y_test_decoded, y_pred_decoded)

# Display confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['MI', 'Normal'], yticklabels=['MI', 'Normal'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Compute and display other classification metrics
accuracy = accuracy_score(y_test_decoded, y_pred_decoded)
precision = precision_score(y_test_decoded, y_pred_decoded, average='weighted')
recall = recall_score(y_test_decoded, y_pred_decoded, average='weighted')
f1 = f1_score(y_test_decoded, y_pred_decoded, average='weighted')

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-Score: {f1:.4f}')
```
