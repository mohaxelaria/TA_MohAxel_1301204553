```python
from scikeras.wrappers import KerasClassifier
```

```python
from sklearn.model_selection import GridSearchCV

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Bidirectional, LSTM, Dense, Dropout, BatchNormalization, Flatten
from tensorflow.keras.regularizers import l1_l2

# Fungsi untuk membuat model dengan parameter yang dapat diatur
def create_model(Bilstm_units=32, Bilstm2_units=64):
    model = Sequential()

    # First Conv1D layer
    model.add(Conv1D(filters=16, kernel_size=3, input_shape=(33, 1), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=2))

    # Second Conv1D layer
    model.add(Conv1D(filters=16, kernel_size=3, activation='relu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.2))

    # BiLSTM Layer 1
    model.add(Bidirectional(LSTM(Bilstm_units, activation='relu', return_sequences=True)))
    model.add(Dropout(0.2))

    # BiLSTM Layer 2
    model.add(Bidirectional(LSTM(Bilstm2_units, activation='relu')))
    model.add(Dropout(0.2))

    # Dense layer
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return model
```

```python
# Parameter grid
param_grid = {
    
    'model__Bilstm_units': [16, 32, 64, 128],
    'model__Bilstm2_units': [16, 32, 64, 128],
    
    'batch_size': [32],
    'epochs': [100]
}
```

```python
X_train_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))
```

```python
# Pembungkus Keras
model = KerasClassifier(model=create_model)
```

```python
# Grid search
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train_reshaped, y_train)
```

```python
# Hasil terbaik
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
